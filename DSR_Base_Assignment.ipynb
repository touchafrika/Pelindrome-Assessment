**Question 2A**

***What is the total number of people living with HIV (NoPLHIV) in the listed districts according to the Survey estimate?***

```python
# Filtering dataset
survey_data = data[data.Estimate.isin(['Survey'])]

# Filtering by relavent columns
total_noplhiv = survey_data[['District', 'Code', 'Estimate', 'NoPLHIV']]

# sorting by NoPLHIV(lowest to highest)
total_noplhiv.sort_values('NoPLHIV')
```

**Question 2B**

***What is the average NoPLHIV of the two estimates used for “Xhariep”?***

```python
# Filtering dataset by Xhariep district 
xhariep_df = data[(data.District.isin(['Xhariep']))]

# calling the Relevent columns
xhariep_df = xhariep_df[['District', 'Code', 'Estimate', 'NoPLHIV']]

# Printing the average of Xhariep District
xhariep_df.mean(axis=1)
```
